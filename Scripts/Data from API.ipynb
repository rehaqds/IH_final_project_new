```python
# Imports

import requests
import pandas as pd
```

```python
# Data USA API
# List of all variables: https://datausa.io/api/cubes
# Ratio of physicians: https://datausa.io/api/data?drilldowns=State&measures=Patient%20to%20Primary%20Care%20Physician%20Ratio&year=2022

url_pollution = 'https://datausa.io/api/data?drilldowns=State&measures=Air%20Pollution&year=2021'
url_population = 'https://datausa.io/api/data?drilldowns=State&measures=Population&Year=2021'
```

```python
 def get_data_from_api(url):   
    # Get the data from the API endpoint
    response = requests.get(url)

    print(response.status_code)
    info = response.json()
    print(info.keys())
    # print(info)
    df = pd.DataFrame(info['data'])
    display(df.head())
    return df
```

```python
# Population API

df = get_data_from_api(url_population)

df_population2021 = df[['ID State', 'State', 'Population']].copy()
df_population2021['ID State'] = df_population2021['ID State'].map(lambda s: s[-2:])
display(df_population2021.sort_values(by='Population', ascending=False))

df_population2021.to_csv('population_2021.csv')
```

```python
# Polution API

df = get_data_from_api(url_pollution)

df_pollution2021 = df[['ID State', 'State', 'Air Pollution']].copy()
df_pollution2021['ID State'] = df_population2021['ID State'].map(lambda s: s[-2:])
display(df_pollution2021.sort_values(by='Air Pollution', ascending=False))

df_pollution2021.to_csv('pollution_2021.csv')
```
